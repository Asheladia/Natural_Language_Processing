# NER Clouds

```python
import spacy
from nltk.corpus import reuters
from spacy import displacy
from wordcloud import WordCloud
import matplotlib.pyplot as plt 

nlp = spacy.load("en_core_web_sm")
```

```python
articles = reuters.raw(fileids = reuters.fileids(categories='gas')[0])
```

```python
doc = nlp(articles)
displacy.render(doc, style = 'ent')
```

```python
articles = reuters.raw(categories='gas')
doc = nlp(articles)
```

```python
entities = [ent.text for ent in doc.ents if ent.label_ in ['GPE', "ORG"]]
```

```python
entities = [i.lower().replace(' ', '_') for i in entities]
entities[0:10]
```

```python
entities = [j for j in entities if (j!='mln' and j!='MLN' and j!='lt')]
```

```python
wc = WordCloud().generate(' '.join(entities))
plt.imshow(wc)
```
